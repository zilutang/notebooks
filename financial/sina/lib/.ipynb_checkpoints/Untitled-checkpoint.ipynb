```python
import pandas as pd
import numpy as np
import seaborn as sns
import urllib
import urllib2
import simplejson
import json
import ast
netamount=lambda x: x['netamount']
opendate=lambda x: x['opendate']
trade=lambda x: x['trade']
jsonloadfun = lambda x: json.loads(x)
stocklist = pd.read_csv('../data/new_jrhy.txt', header=None)
c1=stocklist[0].tolist()
c1 = ['sz002113']
```

```python
ggzjl = r'http://vip.stock.finance.sina.com.cn/quotes_service/api/json_v2.php/MoneyFlow.ssl_qsfx_lscjfb?page=1&num=2000&sort=opendate&asc=0&daima='
```

```python
#rangethis = range(1,2)
#replacestr = lambda strbad: strbad[1].replace('{', '{\"').replace(':\"', '\":\"').replace(',', ',\"').replace('\"{', '{')
#getdata = lambda stockid: (stockid, urllib2.urlopen(urllib2.Request(ggzjl+str(stockid))).read())
replacestr = lambda strbad: strbad.replace('{', '{\"').replace(':\"', '\":\"').replace(',', ',\"').replace('\"{', '{')
getdata = lambda stockid: urllib2.urlopen(urllib2.Request(ggzjl+str(stockid))).read()
tmpresult = map(getdata, c1)
mapedresult = map(replacestr,tmpresult)
jsonresult = map(jsonloadfun, mapedresult)
frame = pd.DataFrame(jsonresult)
frame.to_csv('jsonresult.csv')
```
